<h1><span style="color: #326197">TMVA::Classification</span></h1>

<h1><span style="color: #326197"> Importing required modules</span></h1>

```python
from ROOT import TFile,TMVA
%jsroot on
```

<h1><span style="color: #326197"> Reading the dataset </span></h1>

```python
input = TFile.Open( "dataset.root" )
dl = input.dataset
input.Close()
```

<h1><span style="color: #326197"> Computing Classification</span></h1>
optional you can to call <br>
cl.Train() <br>
cl.Test() <br>
cl.Evaluate() is calling Train() and Test()

```python
cl = TMVA.Classification(dl)
cl.BookMethod(TMVA.Types.kBDT, "BDT","!H:!V:NTrees=850:MinNodeSize=2.5%:MaxDepth=3:BoostType=AdaBoost:nCuts=20")
cl.Evaluate()
```

<h1><span style="color: #326197">Displaying the results</span></h1>

```python
results=cl.GetResults()
results.Print()
results.Draw()
```
